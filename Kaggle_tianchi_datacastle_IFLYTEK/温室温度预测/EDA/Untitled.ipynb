```python
#tem_out空值填充
for i in train_data.index:
    if train_data.loc[i,['tem_out']].isnull()[0]:
        if train_data.loc[i,'hour'] in np.arange(0,7):
            gap_1=(train_data.loc[(train_data['hour']>=0)&(train_data['hour']<7),:].groupby(['hour'])['tem_out'].mean().max()-train_data.loc[(train_data['hour']>=0)&(train_data['hour']<7),:].groupby(['hour'])['tem_out'].mean().min())/25200
            gap_2=train_data.loc[i,'time']-train_data.loc[i-1,'time']
            train_data.loc[i,'tem_out']=train_data.loc[i-1,'tem_out']-(gap_1*gap_2)
        elif train_data.loc[i,'hour'] in np.arange(7,15):
            gap_1=(train_data.loc[(train_data['hour']>=7)&(train_data['hour']<15),:].groupby(['hour'])['tem_out'].mean().max()-train_data.loc[(train_data['hour']>=7)&(train_data['hour']<15),:].groupby(['hour'])['tem_out'].mean().min())/28800
            gap_2=train_data.loc[i,'time']-train_data.loc[i-1,'time']
            train_data.loc[i,'tem_out']=train_data.loc[i-1,'tem_out']+(gap_1*gap_2)
        elif train_data.loc[i,'hour'] in np.arange(15,24):
            gap_1=(train_data.loc[(train_data['hour']>=15)&(train_data['hour']<24),:].groupby(['hour'])['tem_out'].mean().max()-train_data.loc[(train_data['hour']>=15)&(train_data['hour']<24),:].groupby(['hour'])['tem_out'].mean().min())/232400
            gap_2=train_data.loc[i,'time']-train_data.loc[i-1,'time']
            train_data.loc[i,'tem_out']=train_data.loc[i-1,'tem_out']-(gap_1*gap_2)

        
```
